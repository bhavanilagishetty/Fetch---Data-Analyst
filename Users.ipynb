```python
import json
import pandas as pd
from pandas.io.json import json_normalize
from datetime import datetime

# Read the JSON file
with open('D:/Resume/FetchInterview/users.json') as f:
    #data = json.load(f)
    data = f.read()
    data = data.split("}\n")
    data = [d.strip() + "}" for d in data]
    data = list(filter(("}").__ne__, data))
    data = [json.loads(d) for d in data]

df = pd.DataFrame(data)

print(df.info())
print(df.head(2))
# Write to CSV file
#df.to_csv('D:/Resume/FetchInterview/receipts.json - Copy/output1.csv', index=False)
```

```python
def get_id_oid(row):
    return row['$oid']

df['id_oid'] = df['_id'].apply(lambda x: get_id_oid(x))

df.drop(columns=['_id'], inplace=True)
df
```

```python
def extract_date(row):
    if isinstance(row, dict) and '$date' in row:
        return datetime.utcfromtimestamp(row['$date'] / 1000).strftime('%Y-%m-%d %H:%M:%S')
    else:
        return None

df['createdDate1'] = df['createdDate'].apply(extract_date)
df['lastLogin1'] = df['lastLogin'].apply(extract_date)

df.drop(columns=['createdDate', 'lastLogin'], inplace=True)
df
```

```python
# Write to CSV file
df.to_csv('D:/Resume/FetchInterview/users.csv', index=False)
```
